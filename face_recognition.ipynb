```python
# модуль в котором содержатся модели распознавания
import face_rec_mod
# библиотеки
import glob
import dlib
import numpy as np
from skimage import io
from scipy.spatial import distance
import json
import cv2
import face_recognition
import time
from datetime import datetime
```

```python
def databaseParse(databasePath):
    o = open(databasePath)
    desc = []
    name = []
    for line in o:
        desc.append(json.loads(line)['desc'])
        name.append(json.loads(line)['name'])
    o.close()
    return {'desc': desc, 'name': name}
```

```python
#идентификация лица
def FaceCompare (face_desc, descArr, nameArr):
    matches = face_recognition.compare_faces(np.array(descArr), np.array(face_desc))
    face_distance = face_recognition.face_distance(np.array(descArr), np.array(face_desc))
    if True in matches:
        matchedIdxs = [i for (i, b) in enumerate(matches) if b]
        counts = {}
        for i in matchedIdxs:
            name =  nameArr[i]
            counts[name] = face_distance[i]

       # The known encoding with the most number of matches corresponds to the detected face name
        name = min(counts, key=counts.get)
        return name
    else: 
        return 'unknown'
```

```python
# выводим имя пользователя с камеры
def faceRecCam (databasePath, videoPath):
    # Get a reference to webcam
    video_capture = cv2.VideoCapture(videoPath)

    # Initialize variables
    face_locations = []
    names_on_video = []
    font = cv2.FONT_HERSHEY_PLAIN 
    face_name = ''
    data = databaseParse(databasePath)

    while (video_capture.isOpened()):
        # Grab a single frame of video
        ret, frame = video_capture.read()
        if (not ret):
            break
        # Convert the image from BGR color (which OpenCV uses) to RGB color (which face_recognition uses)
        rgb_frame = frame[:, :, ::-1]
        # Find all the faces in the current frame of video
        face_locations = face_recognition.face_locations(rgb_frame)
        
        # Display the results
        for top, right, bottom, left in face_locations:
            encoding = face_recognition.face_encodings(frame, [(top, right, bottom, left)])[0]
            face_name = FaceCompare (encoding, data['desc'], data['name'])
            # Draw a box around the face
            cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)
            cv2.rectangle(frame, (left, bottom - 25), (right, bottom), (0, 0, 255), cv2.FILLED)
            font = cv2.FONT_HERSHEY_DUPLEX
            if (face_name != 'unknown'):
                names_on_video.append(face_name)
            cv2.putText(frame, face_name, (left + 6, bottom - 6), font, 0.5, (255, 255, 255), 1)
            # Display the resulting image
        cv2.namedWindow('Video', cv2.WINDOW_NORMAL)
        cv2.imshow('Video', frame)
            # Hit 'q' on the keyboard to quit!
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        # Release handle to the webcam
    uniq_names = set(names_on_video)
    if (uniq_names != set()):
        print("успешно распознаны  {}.".format(uniq_names))

        f = open('names_on_video.txt', 'w')
        for index in uniq_names:
            f.write(index + '\n')
        f.close()
    
    video_capture.release()
    cv2.destroyAllWindows()
    return names_on_video
```

```python
# распознаем лица на видео с диска или камеры (указать 0 во втором аргументе функции faceRecCam)
start_time = datetime.now()
faceRecCam ('database.txt', 0)
end_time = datetime.now()
print("time {}.".format(end_time - start_time))
```

```python
# проверка на основе датасета Georgia Tech face database
start_time = datetime.now()
face_name = ''
count = 0
for image_file_name in glob.glob('gt_db/s*/*.jpg'):
    face_name = faceRecCam('GTdatabase.txt', image_file_name)
    print("face_name {}: image_file_name: {}".format(face_name, image_file_name[6:-7]))
    if(face_name != []):
        if (face_name[0] != image_file_name[6:-7]):
            count+=1
    else:
        count+=1
end_time = datetime.now()
print("time {}, count {}.".format(end_time - start_time, count))
```
